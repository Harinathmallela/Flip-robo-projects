```python
#lets import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
 

import re
pd.set_option('display.max_row',None)
```

## Question 1

```python
def question1(url):
    driver = webdriver.Chrome("chromedriver.exe") #activating the web driver
    driver.get(url)  #getting the required page
    column_list = ['Rank','Name','Artist','Views (In Billions)','Upload Date','Note']  #creating the columns list
    Top_30_df = pd.DataFrame(columns=column_list, index= list(range(0,30)))  #since there are 30 videos
    delay = 10
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.ID , 'Top_videos')))  #wait till we find this id 
    for i in range(30):   #30 rows
        for j in range(6):  #6 columns
            #using dot format method to nagivate to each cell
            try:
                row = driver.find_element_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[{}]".format(i+1,j+1))
                Top_30_df.iloc[i,j] = row.text  #appending to dataframe
            except NoSuchElementException:
                Top_30_df.iloc[i,j] = '-'
    Top_30_df.drop('Note',axis = 1,inplace= True) #not required
    driver.close()
    return Top_30_df  #returning dataframe as result
```

## Question 2

```python
def question2(url):
    driver = webdriver.Chrome("chromedriver.exe") #activating the chrome browzer
    driver.get(url)  #getting the url
    
    #waiting till we find the below xpath
    delay = 10
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH , "//a[@class = 'navigation__link navigation__link--in-drop-down']")))  #wait till we find this xpath

    #getting the link of international fixtures
    international_fixtues = driver.find_element_by_xpath("//a[@class = 'navigation__link navigation__link--in-drop-down']")
    international_fixtues_url = international_fixtues.get_attribute('href')
    driver.get(international_fixtues_url)

    
    #tittle
    time.sleep(5)
    match_title_list = []
    match_title = driver.find_elements_by_xpath("//p[@class = 'fixture__additional-info']/strong")
    for i in match_title:
        match_title_list.append(i.text)
    
    #place
    match_location_list = []
    match_loaction = driver.find_elements_by_xpath("//p[@class = 'fixture__additional-info']/span")
    for i in match_loaction:
        match_location_list.append(i.text)
    
    #series
    series_list = []
    series = driver.find_elements_by_xpath("//span[@class = 'u-unskewed-text fixture__tournament-label u-truncated']")
    for i in series:
        series_list.append(i.text)
    
    #date
    dates_list = []
    months_list= []
    dates = driver.find_elements_by_xpath("//span[@class ='fixture__date']")
    months = driver.find_elements_by_xpath("//span[@class ='fixture__month']")
    for i in dates:
        dates_list.append(i.text)
    for j in months:
        months_list.append(j.text)
    date_month = []
    for i in range(len(dates_list)):
        x = dates_list[i] + 'th' + ' ' + months_list[i]  #creating a new list which contains both date and month
        date_month.append(x)
    
    #time
    times_list = []
    times = driver.find_elements_by_xpath("//span[@class ='fixture__time']")
    for i in times:
        times_list.append(i.text)
    
    #teams
    teams_list = []
    teams = driver.find_elements_by_xpath("//div[@class ='fixture__teams']")
    for i in teams:
        teams_list.append(i.text.replace('\n',' '))
    
    #creating a empty dataframe
    international_fixtures_df = pd.DataFrame({})
    
    international_fixtures_df['Match Title'] = match_title_list #creating a column associated with the list
    international_fixtures_df['Series'] =series_list
    international_fixtures_df['Teams'] = teams_list
    international_fixtures_df['Place'] = match_location_list
    international_fixtures_df['Date'] =date_month
    international_fixtures_df['Time'] =times_list
    driver.close()
    return international_fixtures_df #returning dataframe as a result
```

```python
#calling the function to get dataframe as a result
question2("https://www.bcci.tv/")
```

## Question 3

```python
def question3(url):
    driver = webdriver.Chrome("chromedriver.exe") #activating browzer
    driver.get(url)  #getting the url
    
    
    #main page
    delay = 10  #delay time
    #waiting till the element is found
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//li[@class = 'fa fa-chevron-circle-right']")))
    selenium_page = driver.find_elements_by_xpath("//li[@class = 'fa fa-chevron-circle-right']/a") 
    for i in selenium_page:
        if i.text == 'Selenium':  #clicking the button if selenium is present
            i.click()
            break
    
    #selenium page
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//td[@class = 'responsivetable']")))
    selenium_expections_page = driver.find_elements_by_xpath("//td[@class = 'responsivetable']/a")
    for i in selenium_expections_page:
        title = i.get_attribute('title')
        if 'Selenium Exception Handling' in title:  #clicking the button if exception handling is present
            link = i.get_attribute('href')
            driver.get(link)
            break
    
    #selenium exceptions page
    exceptions_df = pd.DataFrame({})  #creating the empty dataframe
    mylist = ['Exceptions','Description']  #creating columns list
    for i in range(1,len(mylist)+1):
        temp_list = []  #creating temporary list
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH , "//table[@class = 'table table-striped']/tbody")))
        exceptions = driver.find_elements_by_xpath("//table[@class = 'table table-striped']/tbody/tr/td[{}]".format(i))
        for j in exceptions:
            try:
                temp_list.append(j.text)
            except:
                temp_list.append('-')
        exceptions_df[mylist[i-1]] = temp_list   #creating columns in dataframe associated with the list
    exceptions_df.drop(0, axis = 0 ,inplace = True)
    
    driver.close()
    return exceptions_df   #returing the dataframe as a result
```

## Question 4

```python
def question4(url):

    driver = webdriver.Chrome("chromedriver.exe")  #activating browser
    driver.get(url)   #getting url

    delay = 10  #waiting till we found the elememt to be found
    
    
    #main page
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//div[@class = 'dropdown-content']")))
    economy_dropdown = driver.find_elements_by_xpath("//div[@class = 'dropdown-content']/a") #dropdrop menu
    for i in economy_dropdown:
        try:
            indian_economy = i.get_attribute('href')  #getting the link
            if 'economy/india' in indian_economy:
                driver.get(indian_economy)   #moving to next page from the link
                break
        except StaleElementReferenceException:
            pass

    #second page
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//a[@class = 'ec']")))
    indian_states_gdp =driver.find_elements_by_xpath("//a[@class = 'ec']")  #finding the path
    for i in indian_states_gdp:
        try:
            if "GDP of Indian states" in i.text:
                link = i.get_attribute('href')  #getting the link
                driver.get(link)
                break
        except:
            pass
    
    
    #creating the columns name list
    columns_list = ['Rank','State','GSDP at current price (19-20)','GSDP at current price (18-19)','Share(18-19)','GDP($ billion)']
    
    #Creating the empty dataframe
    gdp_df = pd.DataFrame({})
    
    #final page containing gdp table
    for i in range(len(columns_list)):
        temp_list = []
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//table[@id ='table_id']/tbody/tr")))
        try:
            #xpath to get all elements in table using dot format method
            path = driver.find_elements_by_xpath("//table[@id ='table_id']/tbody/tr/td[{}]".format(i+1))  
            for j in path:
                try:
                    temp_list.append(j.text)
                except NoSuchElementException:
                    temp_list.append('-')
            gdp_df[columns_list[i]] = temp_list
        except NoSuchElementException:
            pass   
        
    driver.close()
    return gdp_df  #returning the dataframe as result
```

## Question 5

```python
def question5(url):
    driver = webdriver.Chrome("chromedriver.exe") #activating browser
    driver.get(url) #getting url
    
    
    #main page
    delay = 10  #wait till the element is found
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//div[@class = 'dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul/li/a")))
    explore_menu = driver.find_elements_by_xpath("//div[@class = 'dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul/li/a")
    for i in explore_menu:
        link = i.get_attribute('href') #getting the trending page link
        if 'trending' in link:
            driver.get(link)
            break
    
    #trending page
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//h1[@class = 'h3 lh-condensed']/a")))
    links_list = []
    links = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']/a")
    for i in links:  
        links_list.append(i.get_attribute('href')) #getting all the links of trending repositaries


    desc_list = []
    title_list = []
    contributors_list = []
    language_list = []
    for link in links_list: #looping through all the links
        driver.get(link)

        delay = 10
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//div[@class = 'BorderGrid BorderGrid--spacious']/div")))

        #description 
        try:
            desc = driver.find_element_by_xpath("//div[@class = 'BorderGrid-row hide-sm hide-md']/div/p")
            desc_list.append(desc.text)
        except NoSuchElementException:
            desc_list.append('-')

        #contributors
        try:
            contributors = driver.find_elements_by_xpath("//div[@class = 'BorderGrid BorderGrid--spacious']/div")
            mystr = ''
            for i in contributors:
                mystr += i.text
                #using regex to get the required data
                if 'Contributors' in i.text:
                    count = re.search(r'Contributors\s([\d]+)',i.text).group().replace('Contributors','')
                    contributors_list.append(count)
            if 'Contributors' not in mystr:
                contributors_list.append('-')
        except:
            contributors_list.append('-')
        
        #languages used in the repositary
        try:
            languages = driver.find_elements_by_xpath("//div[@class = 'BorderGrid BorderGrid--spacious']/div")
            mystr = ''
            for i in languages:
                mystr += i.text
                if 'Languages' in i.text:
                    lang = i.text.replace('Languages','')
                    language_list.append(lang.replace('\n', ' '))     
            if 'Languages' not in mystr:
                language_list.append('-')
        except:
            language_list.append('-')

        #title of repositary
        try:
            title = driver.find_element_by_xpath("//h1[@class = ' d-flex flex-wrap flex-items-center break-word f3 text-normal']")
            title_list.append(title.text.replace('\n',''))
        except NoSuchElementException:
            title_list.append('-')

    trending_rep_df = pd.DataFrame({}) #creating the empty dataframe
    trending_rep_df['Repository Title'] = title_list  #creating the column associated with list
    trending_rep_df['Description'] = desc_list
    trending_rep_df['Contributors Count'] = contributors_list
    trending_rep_df['Languages Used'] = language_list
    
    driver.close()
    return trending_rep_df    #returing dataframe as a result
```

## Question 6

```python
def question6(url):
    
    driver = webdriver.Chrome("chromedriver.exe")  #activating browser
    driver.get(url)  #getting url
    
    #main page
    delay = 10  #wait till we find this xpath
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//ul[@class = 'header__submenu__list']/li")))
    charts = driver.find_elements_by_xpath("//ul[@class = 'header__submenu__list']/li/a")
    for i in charts:
        link = i.get_attribute('href')
        if 'charts/hot-100' in link:  #getting link of required page
            driver.get(link)
            break



    songs_df  = pd.DataFrame({}) #creating empty dataframe
    col_list = ['Song Name' ,'Artist Name']  #creating column names
    
    #Hot-100 songs page
    for i in range(len(col_list)): #getting artist and song name
        temp_list = []
        song_artist = driver.find_elements_by_xpath("//span[@class ='chart-element__information']/span[{}]".format(i+1))
        for j in song_artist:
            temp_list.append(j.text)
        songs_df[col_list[i]] = temp_list 
    
    #lastweek rank
    lastweek_list = []
    lastweek = driver.find_elements_by_xpath("//div[@class = 'chart-element__meta text--center color--secondary text--last']")
    for i in lastweek:
        lastweek_list.append(i.text)
    
    #peak rank
    peak_list = []
    peaks = driver.find_elements_by_xpath("//div[@class = 'chart-element__meta text--center color--secondary text--peak']")
    for i in peaks:
        peak_list.append(i.text)
    
    #weeks on board
    weeks_on_board_list = []
    weeks_on_board = driver.find_elements_by_xpath("//div[@class = 'chart-element__meta text--center color--secondary text--week']")
    for i in weeks_on_board:
        weeks_on_board_list.append(i.text)

    songs_df['Last Week Rank'] = lastweek_list #creating the column associated with list
    songs_df['Peak Rank'] = peak_list
    songs_df['Weeks on Board'] =weeks_on_board_list
    
    driver.close()
    return songs_df  #returning dataframe as result
```

```python

def question7(url):
    driver = webdriver.Chrome('chromedriver.exe') #activating browser

    driver.get(url) #getting url

    delay = 10  #wating till the elements is found
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//ul[@class = 'midSec menu']")))
    recruiters_page = driver.find_element_by_xpath("//ul[@class = 'midSec menu']/li[2]/a")
    recruiters_page_link = recruiters_page.get_attribute('href')
    driver.get(recruiters_page_link)  #getting recruiter page link

    #searching for data science in recruiter page
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//div[@class = 'inpWrap']/input")))
    input_tab = driver.find_element_by_xpath("//div[@class = 'inpWrap']/input")
    input_tab.send_keys('data science')  #sending input
    time.sleep(2)
    search_button = driver.find_element_by_xpath("//button[@class = 'fl qsbSrch blueBtn']")
    search_button.click()

    #getting the profile links of all recruiters
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//p[@class ='highlightable']/a")))
    links_list = []
    links = driver.find_elements_by_xpath("//p[@class ='highlightable']/a[1]")
    for i in links:
        links_list.append(i.get_attribute('href'))

    name_list = []
    designation_list = []
    company_list = []
    location_list = []
    skills_list = []
    for link in links_list: #looping in through all the links of recruiters
        driver.get(link)

        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//h1[@class = 'fl ellipsis wLimit hd']")))
        
        #name
        try:
            names= driver.find_elements_by_xpath("//h1[@class = 'fl ellipsis wLimit hd']")
            for i in names:
                name_list.append(i.text)
        except NoSuchElementException:
            name_list.append('-')

        #recruiter designation
        try:
            designations = driver.find_elements_by_xpath("//div[@class = 'rFrame fl infoWrapper']/div[3]")
            for i in designations:
                designation_list.append(i.text)
        except NoSuchElementException:
            designation_list.append('-')

        #company and location
        try:
            company_location = driver.find_elements_by_xpath("//a[@class = 'fl ellipsis widLrg']")
            for i in company_location[:1]:
                company_list.append(i.text)
            if len(company_location) == 2:
                for j in company_location[1:2]:  #since both are same xpaths, looping in 2 varibales
                    location_list.append(j.text)
            else:
                location_list.append('-')
        except NoSuchElementException:
            company_list.append('-')
            location_list.append('-')

        #skills required
        try:
            skills = driver.find_elements_by_xpath("//div[@class = 'fl lPortn']/p")
            mystr = ''
            for i in skills:
                mystr += i.text
            skills_list.append(mystr)
        except NoSuchElementException:
            skills_list.append('-')
        
        
            
    HR_df = pd.DataFrame({})  #creating a empty dataframe
    HR_df['Name'] = name_list   #creating columns associated with the list
    HR_df['Desigantion'] =designation_list
    HR_df['Company'] =company_list
    HR_df['Locations'] =location_list
    HR_df['Skills Required'] =skills_list
    
    driver.close()
    return HR_df #returing the dataframe as a result
```

## Question 8

```python
def question8(url):
    driver = webdriver.Chrome('chromedriver.exe') #activating the browser
    driver.get(url) #getting url
    
    
    delay = 10 #waiting till the elememt is found
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,'//tbody/tr/td')))
    
    #getting columns
    columns_list = []
    columns = driver.find_elements_by_xpath('//thead/tr/th')
    for i in columns:
        columns_list.append(i.text)
        
    #getting table content
    table_list = []
    table = driver.find_elements_by_xpath('//tbody/tr/td')
    for i in table:
        table_list.append(i.text)
    
    df_books = pd.DataFrame({}) #creating the empty dataframe
    for i in range(len(columns_list)):
        temp_list = table_list[i::len(columns_list)]  #slicing to get required data
        df_books[columns_list[i]] = temp_list
        
    driver.close()
    return df_books  #returning the dataframe as result
```

```python
def question9(url):
    driver = webdriver.Chrome('chromedriver.exe') #activating browser
    driver.get(url) #getting url
    
    delay = 10 #waiting till the elememt is found
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//div[@class = 'ipl-rating-star small']/span[2]")))
    
    #getting xpaths of all columns and making a dictionary
    xpath_dict = {'Name' : "//h3[@class = 'lister-item-header']/a", 'Years Span' : "//h3[@class = 'lister-item-header']/span[2]",
                 'Run Time' : "//div[@class = 'lister-item-content']/p[1]/span[3]" ,'Genre' : "//div[@class = 'lister-item-content']/p[1]/span[5]",
                 'Ratings': "//div[@class = 'ipl-rating-star small']/span[2]" , 'Votes':"//div[@class = 'lister-item-content']/p[4]/span[2]" }

    top_shows_df = pd.DataFrame({})  #creating a empty dataframe
    for i in xpath_dict.items():  #items have keys and values
        temp_list = []
        temps = driver.find_elements_by_xpath(i[1])  #getting xpaths
        for j in temps:
            temp_list.append(j.text)
        top_shows_df[i[0]] = temp_list   #creating columns in dataframe
    driver.close()
    return top_shows_df   #returning in the form of a dataframe
```

```python
def question10(url):
    
    driver = webdriver.Chrome('chromedriver.exe') #activating chrome driver
    driver.get(url)
    
    delay = 10 #waiting till the element is found
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//span[@class ='normal']/b/a")))
    all_datasets = driver.find_element_by_xpath("//span[@class ='normal']/b/a")
    all_datasets.click()  #clicking on all datasets link
    
    #creating the columns list
    column_names = ['Dataset Name', 'Datatype' ,' Task' ,'Atrribute Type' , 'Number of Instances' , 'Number of Atrributes' , 'Year']

    #getting table content
    table_list = []
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//table/tbody/tr/td[2]/table[2]/tbody/tr/td")))
    table = driver.find_elements_by_xpath('//table/tbody/tr/td[2]/table[2]/tbody/tr/td')
    for i in table:
        table_list.append(i.text)  #getting the entire table content into a list

    dataset_df = pd.DataFrame({}) #creating the empty dataframe
    for i in range(len(column_names)):
        temp_list = table_list[i::len(column_names)]  #slicing to get required data
        dataset_df[column_names[i]] = temp_list
        
    driver.close()
    dataset_df.drop(0,axis =0 ,inplace = True)
    return dataset_df  #returing the dataframe as result
```
